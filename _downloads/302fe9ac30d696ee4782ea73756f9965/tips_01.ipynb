
# 3D visualization

Visualize 3D imported geometry


## Import necessary libraries



```python
import ansys.mechanical.core as mech
from ansys.mechanical.core.examples import delete_downloads, download_file
```

Embed mechanical and set global variables



```python
app = mech.App(version=242)
app.update_globals(globals())
print(app)
```

## Download and import geometry
Download geometry



```python
geometry_path = download_file("Valve.pmdb", "pymechanical", "embedding")
```

Import geometry



```python
geometry_import = Model.GeometryImportGroup.AddGeometryImport()
geometry_import.Import(geometry_path)
```

## Visualize in 3D



```python
app.plot()
```

<div class="alert alert-info"><h4>Note</h4><p>This visualization is currently available only for geometry and on version 24R2 or later</p></div>



## Cleanup



```python
delete_downloads()
app.new()
```
